```python
mlr_v80_text = scrap.read_proceedings(mlr_v80)
```

```python
mlr_v80_text.head()
```

```python
# Check your regex here: https://pythex.org/
regexes = {
    'f1': r'F[1 -]*?(measure|score)',
    'auc': r'AU(C|ROC|CH)',
    'logloss': r'[Ll]og[ -]*?loss',
    'acc': r'accuracy',
    'imbalance': r'(im|un|not )balance'
}
```

## RegEx on a subset ##

```python
selected_regexes = scrap.subdict(regexes, ['f1', 'auc'])
selected_regexes_counts = scrap.regexes_in_proceedings(selected_regexes, mlr_v80_text)
selected_regexes_counts.head()
```

```python
scrap.regexes_to_matrix(selected_regexes_counts)
```

## RegEx on all (and subset) ##

```python
regexes_counts = scrap.regexes_in_proceedings(regexes, mlr_v80_text)
regexes_counts.head()
```

```python
scrap.regexes_to_matrix(regexes_counts)
```

```python
scrap.regexes_to_matrix(regexes_counts, regexes_in=['f1', 'auc', 'logloss'])
```
